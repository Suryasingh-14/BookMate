````python
books.rename(columns={'Book-Title':'title',
                       'Book-Author':'author',
                       'Year-Of-Publication':'year',
                          'Publisher':'publisher',
                        'Image-URL-L':'image_url'}, inplace=True)
````

````python
# Now Load the second dataset (USERS INFO)
users = pd.read_csv('users.csv', sep=';', on_bad_lines='skip', encoding='latin-1')
````

````python
users.head()
````

````python
# Rename columns for better usability
users.rename(columns={'User-ID':'user_id',
                       'Location':'location',
                       'Age':'age'}, inplace=True)
````

````python
users.head()
````

````python
#Now Load the third Dataset (BOOK RATINGS)
ratings = pd.read_csv('ratings.csv', sep=';',on_bad_lines='skip',encoding='latin-1')
````

````python
ratings.head()
````

````python
#Let's rename columns for better usability
ratings.rename(columns={'User-ID':'user_id',
                        'Book-Rating':'rating'}, inplace=True)
````

````python
ratings.head()
````

````python
#Let's store that users who at least rated more than 200 books are considered as "active users"
x = ratings['user_id'].value_counts()>=200
````

````python
x[x].shape
````

````python
#Index
y=x[x].index
````

````python
y
````

````python
#Filter the ratings dataframe to include only active users
ratings = ratings[ratings['user_id'].isin(y)]
````

````python
ratings.head()
````

````python
# Merge Ratings with Books
ratings_with_books = ratings.merge(books, on='ISBN')
````

````python
ratings_with_books.head()
````

````python
# which book has which rating
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
````

````python
number_rating.head()
````

````python
number_rating.rename(columns={'rating':'num_of_ratings'},inplace=True)
````

````python
number_rating.head()
````

````python
#merge number_rating with ratings_with_books
final_ratings = ratings_with_books.merge(number_rating, on='title')
````

````python
final_ratings.head()
````

````python
#Filter those books who got more than 50 ratings
final_ratings = final_ratings[final_ratings['num_of_ratings']>=50]
````

````python
# Let's drop duplicates if any
final_ratings.drop_duplicates(['title', 'user_id'],inplace=True)
````

````python
final_ratings.head()
````

````python
final_ratings.isnull().sum()
````

````python
#Let's create a pivot table with users and ratings
book_pivot = final_ratings.pivot_table(columns='user_id', index='title', values='rating')
````

````python
book_pivot.head()
````

````python
book_pivot.fillna(0,inplace=True)
book_pivot.head()
````

````python
from scipy.sparse import csr_matrix
````

````python
book_sparse=csr_matrix(book_pivot)
book_sparse
````

````python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
````

##  What is `sklearn.neighbors`?
`sklearn.neighbors` scikit-learn ka module hai jo similarity-based algorithms provide karta hai.
Iska use recommendation systems me similar users/items dhoondne ke liye hota hai.

---

##  What is `NearestNeighbors`?
`NearestNeighbors` ek model hai jo kisi item ke sabse similar (nearest) items find karta hai.
Collaborative Filtering me:
- Similar users find karne ke liye
- Similar books find karne ke liye

---

##  Why use `NearestNeighbors` in Collaborative Filtering?
CF ka idea simple hai:
**"Similar users  similar taste"**  
**"Similar books  similar ratings"**

So hame similarity calculate karni hoti hai  NearestNeighbors yeh kaam karta hai.

---

##  What is `algorithm='brute'`?
`brute` matlab brute-force search:
- Har item ko har item se compare karta hai
- High-dimensional + sparse data ke liye best
- Rating matrices hamesha sparse hoti hain  isliye brute best choice hai

Other algorithms (kd_tree, ball_tree) CF ke liye kaam nahi karte.

---

Abhi ye use nahi kiya 
##  Final Code
```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(
    algorithm='brute',   # brute force similarity search
    metric='cosine'      # cosine similarity for recommendations
)
```

Cosine similarity recommendation systems me sabse accurate hoti hai.



````python
#Fit data
model.fit(book_sparse)
````
